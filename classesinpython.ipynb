```python
class Dog: #class Dog: This tells Python we are starting a new blueprint named Dog. (By convention, class names start with a capital letter).
    
    # 1. The __init__ method (The Setup) 
    def __init__(self, name, breed): #__init__: This is a special, built-in method called a constructor. It runs automatically the exact moment you create a new object. It sets up the initial state (the attributes) of the object.
        self.name = name    # Attribute
        self.breed = breed  # Attribute

    # 2. A Method (Behavior)
    def bark(self):
        return f"{self.name} says Woof!"
    
    def fetch(self, item):
        return f"{self.name} brought back the {item}."

# --- Using the Blueprint ---

# Creating two distinct "Dog" objects (instantiation)
dog1 = Dog("Buddy", "Golden Retriever")
dog2 = Dog("Lucy", "Poodle")

# Accessing attributes
print(dog2.name)    # Output: Buddy
print(dog2.breed)   # Output: Poodle

# Using methods
print(dog2.bark())          # Output: Buddy says Woof!
print(dog1.fetch("stick"))  # Output: Lucy brought back the stick.
```
